```python
def brand_mandatory_rev(df):
    rowA = []
    rowB = []
    rowC = []
    rowD = []

    for idx_row, row in df.iterrows():
        car_brand_finding = row['carbrand']
        car_model_finding = row['car_model']
        model_th_finding = row['model_name_th']
        nickname_finding = row['nickname']

        car_brand_check = car_brand['brand']
        
        car_brandcheck = []
        for brand in  car_brand_check:
            if brand == car_brand_finding:
                car_brandcheck.append(True)
            else:
                car_brandcheck.append(False)
        found = any(j == True for j in car_brandcheck)
        rowA.append(found)
        carA = car_brand
        if found:
            carA = carA[car_brandcheck]
#         print(carA)
#         print(rowA)

        car_model_check = carA['model']
    #         print(carnametest)
        car_modelcheck = []
        for model in car_model_check:
    #             print(i)
            if model == car_model_finding:
    #                 print(i)
                car_modelcheck.append(True)
            else:
                car_modelcheck.append(False)
        found1 = any(a == True for a in car_modelcheck)
        rowB.append(found1)
        carB = carA
        if found1:
            carB = carB[car_modelcheck]
#         print(car)
#         print(rowB)

        model_th_check = carB['model_th']
        modelth_check = []
        for modelth in model_th_check:
            if modelth == model_th_finding:
                modelth_check.append(True)
            else:
                modelth_check.append(False)
        found2 = any(b == True for b in modelth_check)
        rowC.append(found2)
        carC = carB
        if found2:
            carC = carC[modelth_check]
            
    #     print(car)
    # print(rowC)

        if 'name' not in carC:
            print(f"carA:{carA}")
            print(f"carB:{carB}")
            print(f"carC:{carC}")
            print(f"idx_row: {idx_row}")
            print(f"row: {row}")
            print(f"modelth_check: {modelth_check}")
        nicknamecheck = carC['name']
        
        namecheck = []
        for cname in nicknamecheck:
            if cname == nickname_finding:
                namecheck.append(True)
            else:
                namecheck.append(False)
        found3 = any(k == True for k in namecheck)
        rowD.append(found3)
        carD = carC
        if found3:
            carD = carD[namecheck]
            
#         print(car)
#         print(rowD)

    dfA = pd.DataFrame(rowA, columns = ['car_brand'])
    dfB = pd.DataFrame(rowB, columns = ['car_model'])
    dfC = pd.DataFrame(rowC, columns = ['model_name_th'])
    dfD = pd.DataFrame(rowD, columns = ['nickname'])
    ret = pd.concat([dfA,dfB,dfC,dfD], axis=1)
    ret['index'] = range(1, len(ret) + 1)
    ret = ret.set_index('index')
    return ret
```

```python
from re import search
sub = ","
sub2 = "."
```

```python
def cc_mandatory_rev(df6):
    rowJ = []

    for idx_row, row in df6.iterrows():
        cc_check = row['CC']
        if cc_check == "nan":
            rowJ.append(True)
        else:
            if search(sub, cc_check):
                rowJ.append(False)
            else:
                if cc_check.find(sub2) != -1:
                    f = len(cc_check.split('.')[0])
                    b = len(cc_check.split('.')[1])
                    if f ==1 and b <= 2:
                        rowJ.append(True)
                    else:
                        rowJ.append(False)
                else:
                    cc_len = len(cc_check)
                    if cc_len == 1:
                        rowJ.append(True)
                    else:
                        rowJ.append(False)

    dfJ = pd.DataFrame(rowJ, columns = ['CC'])
    dfJ['index'] = range(1, len(dfJ) + 1)
    ret6 = dfJ.set_index('index')
    return ret6
```
