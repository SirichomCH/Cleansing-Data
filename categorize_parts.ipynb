# Convert xlsx to csv

```python
import pandas as pd
```

```python
df = pd.read_excel('mandatory_11.25B.xlsx', 'car_brand')
df = df.iloc[:,:]
df['model'] = df['model'].astype(str)
df['nickname'] = df['nickname'].astype(str)
df.head()
```

```python
# c = df['model'][0]
# c = c.replace(",","").split()
# c
```

```python
count_model = df['model'].apply(lambda x: len(x.replace(",","").split()))
count_nickname = df['nickname'].apply(lambda x: x.startswith("โฉมปี")).astype(int)
total = count_model + count_nickname
df['total'] = total
df = df.sort_values('total', ascending=False)
df.to_csv('mandatory_edited.csv', index=False)
```
